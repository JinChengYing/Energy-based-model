```python
import torch

# 检测是否有可用GPU（返回True表示GPU可用）
gpu_available = torch.cuda.is_available()
print(f"GPU是否可用：{gpu_available}")

# 查看可用GPU数量和显卡信息
if gpu_available:
    gpu_count = torch.cuda.device_count()
    gpu_name = torch.cuda.get_device_name(0)  # 查看第0块GPU的名称
    print(f"可用GPU数量：{gpu_count}")
    print(f"第1块GPU名称：{gpu_name}")
```
